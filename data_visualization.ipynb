```python
#hiển thị số lượng chuyến bay của mỗi hãng
count_airline = data['code_name'].value_counts(ascending=True)
count_airline.plot.barh(color=['brown', 'lightgreen'])

plt.title('số lượng chuyến bay của mỗi hãng hàng không')
plt.xlabel('Số lượng chuyến bay')
plt.ylabel('Hãng hàng không')

plt.show()
```
